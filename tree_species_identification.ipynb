```python
drive_dataset_path = "/content/drive/MyDrive/archive (1)/Tree_Species_Dataset"
local_dataset_path = "/content/tree-species-dataset"

import shutil
import os

if not os.path.exists(drive_dataset_path):
    raise FileNotFoundError(f"❌ Folder not found at: {drive_dataset_path}")

if os.path.exists(local_dataset_path):
    shutil.rmtree(local_dataset_path)

shutil.copytree(drive_dataset_path, local_dataset_path)

print("✅ Dataset copied successfully to:", local_dataset_path)
```

```python
import os
from collections import Counter
import matplotlib.pyplot as plt

dataset_path = '/content/tree-species-dataset'

# List all species (subfolders)
species = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]
print(f"Found {len(species)} species.")

# Count images per species
species_counts = {sp: len(os.listdir(os.path.join(dataset_path, sp))) for sp in species}
print(species_counts)

# Plot distribution
plt.figure(figsize=(12,6))
plt.bar(species_counts.keys(), species_counts.values())
plt.xticks(rotation=90)
plt.title("Number of Images per Species")
plt.show()
```
